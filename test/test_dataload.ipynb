```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def load_and_preprocess(image_path, indexlabel_path, label_path):
    # Load images
    image = Image.open(image_path)
    indexlabel = Image.open(indexlabel_path)
    label = Image.open(label_path)
    
    # Convert to numpy arrays
    image_array = np.array(image)
    indexlabel_array = np.array(indexlabel)
    label_array = np.array(label)
    
    # Normalize image (if it's not already normalized)
    image_array = image_array / 255.0
    
    return image_array, indexlabel_array, label_array

def visualize_data(image, indexlabel, label):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    ax1.imshow(image)
    ax1.set_title('Original Image')
    ax1.axis('off')
    
    ax2.imshow(indexlabel)
    ax2.set_title('Index Label')
    ax2.axis('off')
    
    ax3.imshow(label)
    ax3.set_title('Semantic Label')
    ax3.axis('off')
    
    plt.tight_layout()
    plt.show()

# Example usage
dir = '../../WildScenes_Dataset-61gd5a0t-/data/WildScenes/WildScenes2d/V-01'
image_path = f'{dir}/image/1623377790-818434554.png'
indexlabel_path = f'{dir}/indexLabel/1623377790-818434554.png'
label_path = f'{dir}/label/1623377790-818434554.png'

image, indexlabel, label = load_and_preprocess(image_path, indexlabel_path, label_path)
visualize_data(image, indexlabel, label)
```
